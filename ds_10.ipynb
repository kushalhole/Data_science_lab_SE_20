### Data Cleaning: `daily_df`

```python
# Convert 'date' column to datetime objects
daily_df['date'] = pd.to_datetime(daily_df['date'])

# Replace negative daily_new_cases with 0
daily_df['daily_new_cases'] = daily_df['daily_new_cases'].apply(lambda x: max(0, x))

# Replace negative daily_new_deaths with 0
daily_df['daily_new_deaths'] = daily_df['daily_new_deaths'].apply(lambda x: max(0, x))

print('--- daily_df info() after cleaning ---\n')
daily_df.info()
print('\n--- daily_df describe() after cleaning ---\n')
display(daily_df.describe())
```

### Data Cleaning: `daily_df`

```python
# Convert 'date' column to datetime objects
daily_df['date'] = pd.to_datetime(daily_df['date'])

# Replace negative daily_new_cases with 0
daily_df['daily_new_cases'] = daily_df['daily_new_cases'].apply(lambda x: max(0, x))

# Replace negative daily_new_deaths with 0
daily_df['daily_new_deaths'] = daily_df['daily_new_deaths'].apply(lambda x: max(0, x))

# Replace negative active_cases with 0
daily_df['active_cases'] = daily_df['active_cases'].apply(lambda x: max(0, x))

print('--- daily_df info() after cleaning ---\n')
daily_df.info()
print('\n--- daily_df describe() after cleaning ---\n')
display(daily_df.describe())
```
