```python
import pandas as pd

# Load the Titanic dataset from the available file
try:
    df = pd.read_csv('/content/Titanic-Dataset.csv')
    print("Dataset loaded successfully!")
    display(df.head())
    print("\nDataset Information:")
    print(df.info())
except FileNotFoundError:
    print("Error: The Titanic-Dataset.csv file was not found at /content/Titanic-Dataset.csv.")
    print("Please ensure the file is correctly placed in the /content/ directory or provide the correct path.")
except Exception as e:
    print(f"An unexpected error occurred while loading the dataset: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print('\nDataset Info after loading:')
df.info()
print('\nFirst 5 rows:')
display(df.head())
```

### 1. Overall Survival Count

Let's first see the raw counts of survivors versus non-survivors.

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

### 2. Survival Rate by Sex

Understanding survival rates based on gender is often a key aspect of Titanic analysis.

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='Sex', y='Survived', data=df, palette='magma')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex')
plt.ylabel('Survival Rate')
plt.show()
```

### 3. Survival Rate by Passenger Class (Pclass)

Passenger class is another critical factor. Let's see how it correlates with survival.

```python
plt.figure(figsize=(7, 5))
sns.barplot(x='Pclass', y='Survived', data=df, palette='plasma')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

### 4. Age Distribution with Survival Status

Examining the age distribution for survivors and non-survivors can reveal interesting insights.

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True, bins=30, palette='coolwarm')
plt.title('Age Distribution by Survival Status (0 = Did Not Survive, 1 = Survived)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Did Not Survive', 'Survived'])
plt.show()
```

### Initial Patterns Observed from Visualizations:

*   **Survival Disparity**: A significant portion of passengers did not survive.
*   **Gender Bias in Survival**: Females had a considerably higher survival rate than males, suggesting 'women and children first' was indeed a factor.
*   **Class Matters**: Passengers in higher classes (Pclass 1) had much better survival chances than those in lower classes (Pclass 3), indicating socio-economic status or deck location played a role.
*   **Age's Role**: The age distribution suggests that children (especially very young ones) might have had better survival rates, while the survival rate for older individuals varies and requires deeper analysis.
