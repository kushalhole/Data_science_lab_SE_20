## Probability and Statistics using Python

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### 1. Generate a normal distribution and plot its histogram.

```python
# Generate a normal distribution
# Parameters: mean (loc), standard deviation (scale), number of samples (size)
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)

# Plot the histogram
plt.figure(figsize=(10, 6))
count, bins, ignored = plt.hist(s, 30, density=True, alpha=0.6, color='g')

# Plot the PDF (Probability Density Function) for comparison
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')

plt.title('Histogram of a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

### 2. Compute mean, median, variance, and standard deviation.

```python
# Calculate statistical measures
mean_s = np.mean(s)
median_s = np.median(s)
variance_s = np.var(s)
std_dev_s = np.std(s)

print(f"Mean of the distribution: {mean_s:.4f}")
print(f"Median of the distribution: {median_s:.4f}")
print(f"Variance of the distribution: {variance_s:.4f}")
print(f"Standard Deviation of the distribution: {std_dev_s:.4f}")
```

### 3. Simulate a coin toss experiment and compute probabilities.

```python
# Simulate a coin toss experiment
# 0 for tails, 1 for heads
num_tosses = 10000
coin_tosses = np.random.randint(0, 2, size=num_tosses)

# Calculate the number of heads and tails
num_heads = np.sum(coin_tosses == 1)
num_tails = np.sum(coin_tosses == 0)

# Compute probabilities
prob_heads = num_heads / num_tosses
prob_tails = num_tails / num_tosses

print(f"Total coin tosses: {num_tosses}")
print(f"Number of heads: {num_heads}")
print(f"Number of tails: {num_tails}")
print(f"Probability of heads: {prob_heads:.4f}")
print(f"Probability of tails: {prob_tails:.4f}")
```
