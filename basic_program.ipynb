```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        fact = 1
        for i in range(1, n + 1):
            fact *= i
        return fact

# Example usage:
number = 5
result = factorial(number)
print(f"The factorial of {number} is {result}")

number = 0
result = factorial(number)
print(f"The factorial of {number} is {result}")

number = -3
result = factorial(number)
print(f"The factorial of {number} is {result}")
```

```python
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage:
num_to_display = 7
display_multiplication_table(num_to_display)
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"The number {number} is Even."
    else:
        return f"The number {number} is Odd."

# Example usage:
num1 = 4
print(check_odd_even(num1))

num2 = 7
print(check_odd_even(num2))

num3 = 0
print(check_odd_even(num3))
```

```python
def is_palindrome(number):
    # Convert the number to a string
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        return f"The number {number} is a palindrome."
    else:
        return f"The number {number} is not a palindrome."

# Example usage:
print(is_palindrome(121))
print(is_palindrome(12321))
print(is_palindrome(12345))
print(is_palindrome(1))
print(is_palindrome(0))
```

# Task
Develop a simple rule-based expert system in Python that can make recommendations based on user input, defining a knowledge base with rules, an inference engine, and a user interface to demonstrate its functionality.

## Define Knowledge Base Structure

### Subtask:
Establish a data structure (e.g., a list of dictionaries) to represent the rules for the expert system. Each rule will consist of conditions and an outcome. For instance, we could define rules for recommending an activity based on weather.


**Reasoning**:
The subtask requires initializing an empty Python list named `knowledge_base` to store expert system rules. This `code_block` command directly addresses that requirement.



```python
knowledge_base = []
print("Initialized an empty knowledge base.")
```

**Reasoning**:
The subtask requires adding rules to the `knowledge_base` list. This `code_block` command demonstrates how to add example rules as dictionaries with 'conditions' and 'outcome' keys, fulfilling the instruction.



```python
knowledge_base.append({
    'conditions': {'weather': 'sunny', 'time_of_day': 'morning'},
    'outcome': 'Go for a walk'
})
knowledge_base.append({
    'conditions': {'weather': 'rainy'},
    'outcome': 'Read a book indoors'
})
knowledge_base.append({
    'conditions': {'weather': 'sunny', 'temperature': 'hot'},
    'outcome': 'Go swimming'
})

print(f"Knowledge base now contains {len(knowledge_base)} rules.")
print("Example rules added to the knowledge base.")
```

## Implement Inference Engine

### Subtask:
Develop a Python function that acts as the inference engine. This function will take user inputs (facts) and, by iterating through the rules in the knowledge base, determine a conclusion or recommendation.


**Reasoning**:
To implement the inference engine, I will define a Python function `inference_engine` that iterates through the `knowledge_base`, checks if rule conditions match provided facts, and returns an outcome or a default message if no match is found, as specified in the instructions.



```python
def inference_engine(facts):
    for rule in knowledge_base:
        conditions_met = True
        for condition, value in rule['conditions'].items():
            if condition not in facts or facts[condition] != value:
                conditions_met = False
                break
        if conditions_met:
            return rule['outcome']
    return 'No specific recommendation based on current facts.'

# Example usage:
user_facts1 = {'weather': 'sunny', 'time_of_day': 'morning'}
recommendation1 = inference_engine(user_facts1)
print(f"Based on facts {user_facts1}, recommendation: {recommendation1}")

user_facts2 = {'weather': 'rainy'}
recommendation2 = inference_engine(user_facts2)
print(f"Based on facts {user_facts2}, recommendation: {recommendation2}")

user_facts3 = {'weather': 'sunny', 'temperature': 'hot'}
recommendation3 = inference_engine(user_facts3)
print(f"Based on facts {user_facts3}, recommendation: {recommendation3}")

user_facts4 = {'weather': 'cloudy'}
recommendation4 = inference_engine(user_facts4)
print(f"Based on facts {user_facts4}, recommendation: {recommendation4}")
```

## Create User Input Interface

### Subtask:
Design a simple command-line interface to allow the user to input their observations or answers to questions posed by the expert system.


**Reasoning**:
To create the user input interface, I will define the `get_user_facts` function that prompts the user for conditions like weather, time of day, and temperature, stores their lowercase responses in a dictionary, and then returns these facts.



```python
def get_user_facts():
    facts = {}
    print("\nPlease provide the following information to get a recommendation:")
    weather = input("What is the weather like today (sunny, rainy, cloudy)? ").lower()
    facts['weather'] = weather

    # Only ask for time_of_day if weather is sunny, as per existing rules
    if weather == 'sunny':
        time_of_day = input("What time of day is it (morning, afternoon, evening)? ").lower()
        facts['time_of_day'] = time_of_day

    # Only ask for temperature if weather is sunny, as per existing rules
    if weather == 'sunny':
        temperature = input("What is the temperature like (hot, mild, cold)? ").lower()
        facts['temperature'] = temperature

    return facts

print("Function 'get_user_facts' defined to collect user input.")
```

**Reasoning**:
Now that the `get_user_facts` function is defined, the next step is to combine it with the `inference_engine` to allow the user to interact with the expert system and receive a recommendation based on their live input.



```python
print("\n--- Expert System Recommendation ---")
user_input_facts = get_user_facts()
final_recommendation = inference_engine(user_input_facts)
print(f"\nBased on your input: {user_input_facts}")
print(f"Our recommendation is: {final_recommendation}")
```
