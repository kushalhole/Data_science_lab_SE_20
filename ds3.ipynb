```python
import numpy as np

# Create a 3x3 square matrix
# Modified the last element to make the matrix invertible
matrix_3x3 = np.array([[1, 2, 3],
                       [4, 5, 6],
                       [7, 8, 10]]) # Changed from 9 to 10

print("Modified 3x3 Matrix:")
print(matrix_3x3)

# Compute the determinant of the modified matrix
determinant = np.linalg.det(matrix_3x3)

print("\nDeterminant of the modified matrix:")
print(determinant)
```

```python
import numpy as np

# Define the coefficient matrix A
A = np.array([[2, 3],
              [4, 5]])

# Define the constant vector B
B = np.array([10, 20])

# Solve the system of linear equations
# The function np.linalg.solve(A, B) computes the solution x to the linear system Ax = B.
x_y = np.linalg.solve(A, B)

print("Coefficient matrix A:")
print(A)
print("\nConstant vector B:")
print(B)
print("\nSolution (x, y):")
print(x_y)
```

```python
# Compute the inverse of the modified matrix
inverse_matrix = np.linalg.inv(matrix_3x3)

print("\nInverse of the modified matrix:")
print(inverse_matrix)
```
