```python
import pandas as pd

# Load the Ferrari History Stock Data from the specified CSV file
try:
    ferrari_df = pd.read_csv('/content/Ferrari History Stock Data.csv')
    print("Successfully loaded the Ferrari Stock Data DataFrame. Here are its columns and info:")
    display(ferrari_df.head())
    ferrari_df.info()
except FileNotFoundError:
    print("Error: The file '/content/Ferrari History Stock Data.csv' was not found. Please ensure it's in the correct path.")
    ferrari_df = pd.DataFrame() # Create an empty DataFrame to avoid further errors
except Exception as e:
    print(f"An unexpected error occurred while loading the CSV: {e}")
    ferrari_df = pd.DataFrame()
```

### 1. Filtering specific rows based on conditions

Let's assume the dataset contains a 'Date' column and a 'Close' price column. We will filter for data after a specific date, for example, after '2022-01-01', and where the 'Close' price is above a certain value, say 200. First, we need to convert the 'Date' column to datetime objects and ensure 'Close' is numeric.

```python
if not ferrari_df.empty and 'Date' in ferrari_df.columns and 'Close' in ferrari_df.columns:
    # Convert 'Date' to datetime objects
    ferrari_df['Date'] = pd.to_datetime(ferrari_df['Date'], errors='coerce')

    # Convert 'Close' to numeric, coercing errors to NaN
    ferrari_df['Close'] = pd.to_numeric(ferrari_df['Close'], errors='coerce')

    # Drop rows with NaN values in 'Date' or 'Close' after conversion
    df_cleaned_filter = ferrari_df.dropna(subset=['Date', 'Close'])

    # Filter data after '2022-01-01' and where 'Close' price is > 200
    filtered_ferrari_df = df_cleaned_filter[
        (df_cleaned_filter['Date'] > '2022-01-01') &
        (df_cleaned_filter['Close'] > 200)
    ]

    if not filtered_ferrari_df.empty:
        print("Filtered Ferrari Stock Data (Date > 2022-01-01 and Close > 200) head:")
        display(filtered_ferrari_df.head())
    else:
        print("No stock data found matching the filter criteria (Date > 2022-01-01 and Close > 200).")
else:
    print("Error: 'Date' or 'Close' column not found or DataFrame is empty. Cannot perform filtering.")
```

### 2. Grouping data by categorical variables and computing aggregate statistics

For stock data, it's common to group by year or month. Let's extract the 'Year' from the 'Date' column and group the *original* `ferrari_df` by 'Year'. We will then compute aggregate statistics (sum, mean, count, min, max) for the 'Volume' and 'Close' prices for each year.

```python
if not ferrari_df.empty and 'Date' in ferrari_df.columns and 'Volume' in ferrari_df.columns and 'Close' in ferrari_df.columns:
    # Ensure 'Date' is datetime and extract 'Year'
    ferrari_df['Date'] = pd.to_datetime(ferrari_df['Date'], errors='coerce')
    ferrari_df['Year'] = ferrari_df['Date'].dt.year

    # Convert 'Volume' and 'Close' to numeric
    ferrari_df['Volume'] = pd.to_numeric(ferrari_df['Volume'], errors='coerce')
    ferrari_df['Close'] = pd.to_numeric(ferrari_df['Close'], errors='coerce')

    # Drop rows with NaN in 'Year', 'Volume', or 'Close' for aggregation
    df_for_aggregation = ferrari_df.dropna(subset=['Year', 'Volume', 'Close'])

    if not df_for_aggregation.empty:
        # Group by 'Year' and compute aggregate statistics
        aggregated_ferrari_df = df_for_aggregation.groupby('Year').agg(
            total_volume=('Volume', 'sum'),
            mean_volume=('Volume', 'mean'),
            count_days=('Date', 'count'), # Count of trading days
            min_close=('Close', 'min'),
            max_close=('Close', 'max'),
            mean_close=('Close', 'mean')
        ).reset_index()

        print("Aggregated Ferrari Stock Data by Year:")
        display(aggregated_ferrari_df)
    else:
        print("No valid data for aggregation after cleaning 'Year', 'Volume', or 'Close' columns.")
else:
    print("Error: Required columns ('Date', 'Volume', 'Close') not found or DataFrame is empty. Cannot perform aggregations.")
```
